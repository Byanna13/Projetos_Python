# Matplotlib

- Matplotlib é uma biblioteca python para visualização de dados e geração de graficos, dos mais báicos até os mais avançados;
- Tem uma exelente integração com o jupyter notebook, permitindo  geração d gráficos no próprio jupyter, alem de recursos como zoom;
- Permite exportar graficos para formatos de vetor, pdf, png, jpg, entre varios outras;
- É possivel personalizar o gráfico e a geração pode ser feita através de funções ou mesmo através de POO.

- intalação: pip install matplotlib

```python
# importando a biblioteca
import matplotlib as mpl
```

```python
# importando a biblioteca que facilita a geração dos gráficos
import matplotlib.pyplot as plt
```

```python
# Este comando serve para gerar os gráficos no jupyter
%matplotlib inline
```

```python
plt.plot([1,2,3],[10,5,15])
plt.show()
```

```python
x = [1,2,3]
y = [10,4,2]
plt.plot(x,y)
plt.show()
```

```python
x = [1, 2, 3]
y = [10, 4, 2]
plt.plot(x,y)
plt.xlabel('Nome do Eixo X')
plt.ylabel('Nome do Eixo Y')
plt.title('Titulo do Gráfico')
plt.show()
```

```python
# Label - adiciona um rótulo ao grafico (não aparece, (pq) é opcional)
# Legend() - adciona o rótulo/legenda do gráfico
x = [1, 2, 3]
y = [10, 4, 2]
plt.plot(x, y, label='Legenda do Grafico')
plt.legend()
plt.xlabel('Nome do Eixo X')
plt.ylabel('Nome do Eixo Y')
plt.title('Titulo do Gráfico')
plt.show()
```

```python
# grid() - adiciona as grades do gráfico
x = [1, 2, 3]
y = [10, 4, 2]
plt.plot(x, y, label='Legenda do Grafico')
plt.legend()
plt.xlabel('Nome do Eixo X')
plt.ylabel('Nome do Eixo Y')
plt.title('Titulo do Gráfico')
plt.grid(True)
plt.show()
```

Cores:
r - vermelho
g - verde
b - azul
c - ciano
y - amarelo
m - magenta
k - preto
w - branco

```python
# color='' - muda a cor da linha do gráfico que étas sendo gerado
x = [1, 2, 3]
y = [10, 4, 2]
plt.plot(x, y, label='Legenda do Grafico', color='m')
plt.legend()
plt.xlabel('Nome do Eixo X')
plt.ylabel('Nome do Eixo Y')
plt.title('Titulo do Gráfico')
plt.grid(True)
plt.show()
```

```python
# Inserindo 2 linhas no mesmo gráfico
x = [1, 2, 4]
y1 = [10, 4, 2]
y2 = [5, 10, 2]
plt.plot(x, y1, color='r')
plt.plot(x, y2, color='k')
plt.show()
```

```python
# Inserindo legenda ao gráfico com as 2 linhas mudando a cor
x = [1, 2, 4]
y1 = [10, 4, 2]
y2 = [5, 10, 2]
plt.plot(x, y1, label='Resultado 1', color='g')
plt.plot(x, y2, label='Resultado 2', color='c')
plt.legend()
plt.grid(True)
plt.show()
```

# Graficos de Barras

```python
x = [1, 2, 3]
y1 = [10, 4, 2]
plt.bar(x,y)
plt.show()
```

```python
x = [1, 2, 3]
y1 = [10, 4, 2]
y2 = [4, 2, 5]
plt.bar(x, y1, label='Legenda do Grafico 1', color='m')
plt.bar(x, y2, label='Legenda do Grafico 2', color='c')
plt.legend()
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Titulo do Gráfico')
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3]

x2 = [(x + 0.2) for x in x]
x3 = [(x - 0.2) for x in x]

y1 = [10, 4, 10]
y2 = [4, 2, 5]

plt.bar(x2, y1, label='Legenda do Grafico 1', color='m', width=0.4)
plt.bar(x3, y2, label='Legenda do Grafico 2', color='c', width=0.4)
plt.legend()
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Titulo do Gráfico')
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3]

x2 = [1.2, 2.2, 3.2]
x3 = [0.8, 1.8, 2.8]
# x2 = [(x + 0.2) for x in x]
# x3 = [(x - 0.2) for x in x]

y1 = [10, 4, 10]
y2 = [4, 2, 5]

plt.bar(x2, y1, label='Legenda do Grafico 1', color='m', width=0.4)
plt.bar(x3, y2, label='Legenda do Grafico 2', color='c', width=0.4)
plt.legend()
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Titulo do Gráfico')
plt.grid(True)
plt.show()
```

# Grafico de Pizza

```python
opcoes = ['Péssimo', 'Ruim', 'Neutro', 'Bom', 'Exelente']
votos = [0.1, 2, 2, 8, 5]
plt.pie(votos, labels=opcoes)
plt.show()
```

```python
opcoes = ['Péssimo', 'Ruim', 'Neutro', 'Bom', 'Exelente']
votos = [0.1, 2, 2, 8, 5]
cores = ['b','r','y','c','coral']
plt.pie(votos, labels=opcoes, colors=cores)
plt.legend()
plt.show()
```

```python
opcoes = ['Péssimo', 'Ruim', 'Neutro', 'Bom', 'Exelente']
votos = [0.1, 2, 2, 8, 5]
cores = ['b','r','y','c','coral']
plt.pie(votos, labels=opcoes, colors=cores, autopct='%1.1f%%')
plt.legend()
plt.show()
```

```python
opcoes = ['Péssimo', 'Ruim', 'Neutro', 'Bom', 'Exelente']
votos = [0.1, 2, 2, 8, 5]
cores = ['b','r','y','c','coral']
plt.pie(votos, labels=opcoes, colors=cores, autopct='%1.1f%%', explode=(0,0,0,0,0.2))
plt.legend()
plt.show()
```

```python
opcoes = ['Péssimo', 'Ruim', 'Neutro', 'Bom', 'Exelente']
votos = [0.1, 2, 2, 8, 5]
cores = ['b','r','y','c','coral']
plt.pie(votos, labels=opcoes, colors=cores, autopct='%1.1f%%', explode=(0,0,0,0.2,0))
plt.legend()
plt.show()
```
